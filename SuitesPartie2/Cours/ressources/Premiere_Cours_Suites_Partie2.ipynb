## Suites de limite finie

```python
def suite1(n):
    for k in range(n + 1):
        yield 1 + (-1) ** k  / (k + 1)
        
def suite2(n):
    for k in range(n + 1):
        yield 1 + 1 / (k + 1)
        
def suite3(n):
    for k in range(n + 1):
        yield 1 - 1 / (k + 1)
```

```python
fig, (ax1, ax2,ax3) = plt.subplots(1,3,sharey=True)

ax1.grid()
ax1.axhline(0,color='black')
ax1.axvline(0,color='black')
ax1.set_xlabel(r'$n$')
ax1.set_ylabel(r'$u_{n}$')
ax1.set_title(r'$u_{n}=1+\frac{(-1)^n}{n} \quad \lim_{n \to +\infty}u_{n}=1$')
ax1.scatter(np.arange(0, 51), [terme for terme in suite1(50)] )

ax2.grid()
ax2.axhline(0,color='black')
ax2.axvline(0,color='black')
ax2.set_xlabel(r'$n$')
ax2.set_ylabel(r'$u_{n}$')
ax2.set_title(r'$u_{n}=1+\frac{1}{n} \quad \lim_{n \to +\infty}u_{n}=1$')
ax2.scatter(np.arange(0, 51), [terme for terme in suite2(50)] )

ax3.grid()
ax3.axhline(0,color='black')
ax3.axvline(0,color='black')
ax3.set_xlabel(r'$n$')
ax3.set_ylabel(r'$u_{n}$')
ax3.set_title(r'$u_{n}=1-\frac{1}{n} \quad \lim_{n \to +\infty}u_{n}=1$')
ax3.scatter(np.arange(0, 51), [terme for terme in suite3(50)] )

fig.savefig('limite-finie.pdf')
```

## Suite de limite infinie

```python
def suite4(n):
    for k in range(n + 1):
        yield k + 10 * np.cos(k)
        
def suite5(n):
    for k in range(n + 1):
        yield -k + 10 * np.cos(k)
        
def suite6(n):
    for k in range(n + 1):
        yield np.sqrt(np.sqrt(k))
        
def suite7(n):
    for k in range(n + 1):
        yield - k ** 2        
```

```python
fig, ax1 = plt.subplots()

ax1.grid()
ax1.axhline(0,color='black')
ax1.axvline(0,color='black')
ax1.set_xlabel(r'$n$', fontdict = {'fontsize' : 15})
ax1.set_ylabel(r'$u_{n}$', fontdict = {'fontsize' : 15})
ax1.set_title(r'$u_{n}=n + 10 \cos(n) \quad \lim_{n \to +\infty}u_{n}=+\infty$', fontdict = {'fontsize' : 15})
ax1.scatter(np.arange(0, 101), [terme for terme in suite4(100)] )


fig.savefig('nuage1-limite-infinie.pdf')
```

```python
fig, ax1 = plt.subplots()

ax1.grid()
ax1.axhline(0,color='black')
ax1.axvline(0,color='black')
ax1.set_xlabel(r'$n$', fontdict = {'fontsize' : 15})
ax1.set_ylabel(r'$u_{n}$', fontdict = {'fontsize' : 15})
ax1.set_title(r'$u_{n}= -n + 10 \cos(n) \quad \lim_{n \to +\infty}u_{n}=-\infty$', fontdict = {'fontsize' : 15})
ax1.scatter(np.arange(0, 101), [terme for terme in suite5(100)] )


fig.savefig('nuage2-limite-infinie.pdf')
```

## Suites sans limite

```python
def suite8(n):
    for k in range(n + 1):
        yield (-1) ** k    
        
def suite9(n):
    for k in range(n + 1):
        yield (-1) ** k   * k
```

```python
fig, ax1 = plt.subplots()

ax1.grid()
ax1.axhline(0,color='black')
ax1.axvline(0,color='black')
ax1.set_xlabel(r'$n$', fontdict = {'fontsize' : 15})
ax1.set_ylabel(r'$u_{n}$', fontdict = {'fontsize' : 15})
ax1.set_title(r'$u_{n}= (-1)^n$', fontdict = {'fontsize' : 15})
ax1.scatter(np.arange(0, 21), [terme for terme in suite8(20)] )


fig.savefig('nuage1-sans-limite.pdf')
```

```python
fig, ax1 = plt.subplots()

ax1.grid()
ax1.axhline(0,color='black')
ax1.axvline(0,color='black')
ax1.set_xlabel(r'$n$', fontdict = {'fontsize' : 15})
ax1.set_ylabel(r'$u_{n}$', fontdict = {'fontsize' : 15})
ax1.set_title(r'$u_{n}= (-1)^{n}n$', fontdict = {'fontsize' : 15})
ax1.scatter(np.arange(0, 21), [terme for terme in suite9(20)] )


fig.savefig('nuage2-sans-limite.pdf')
```

## Motifs géométriques

```python
from turtle import *
from math import sqrt
import os


def sauve(nomFichier):
	"""sauve le canevas contenant le tracé de la sortie
	sous la forme d'un fichier eps et d'un fichier pdf
	de nom nomFichier"""
	#on récupère l'écran de la tortue
	ts = getscreen()
	#on récupère la canevas (zone de tracé) de l'écran de la tortue c'est un canevas
	#auquel on peut appliquer les méthodes des canevas du module Tkinter
	#en particulier on peut le convertir en fichier postscript 
	ts.getcanvas().postscript(file="{0}.eps".format(nomFichier))
	#avec la méthode system du  module os on peut appliquer
	#la commande système epstopdf pour convertir le ficher pstscript
	#précédent en fichier pdf
	os.system("epstopdf {0}.eps".format(nomFichier))
    
def repere(xmin, xmax, xgrad, ymin, ymax, ygrad):
    ##----- Fenêtre d'affichage et options -----##
    fenetre = Screen()								# Instance de la fenêtre
    fenetre.setworldcoordinates(xmin, ymin, xmax, ymax)
    speed(0)										# Vitesse la plus rapide
    ##----- Axe des abscisses -----##
    up()
    goto(xmin, 0)
    down()
    goto(xmax, 0)
    nbxgrad = int((xmax - xmin)/xgrad  + 1)
    long_xgrad = (ymax - ymin)/400.0 * 5
    for i in range(nbxgrad):
        x = xmin + i * xgrad
        up()
        goto(x, -long_xgrad)
        down()
        goto(x, long_xgrad)
    ##----- Axe des ordonnées -----##
    up()
    goto(0, ymin)
    down()
    goto(0, ymax)
    nbygrad = int((ymax - ymin)/ygrad  + 1)
    long_ygrad = (xmax - xmin)/400.0 * 5
    for i in range(nbygrad):
        y = ymin + i * ygrad
        up()
        goto(-long_ygrad, y)
        down()
        goto(long_ygrad, y)   
        
def triangle(hypo, n):    
    forward(hypo)
    width(2)
    color("red")    
    write("A({})".format(n))
    left(135)    
    forward(hypo/2*sqrt(2))
    width(1)
    color("blue")
    left(90)
    forward(hypo/2*sqrt(2))
    
    
```

```python
def spirale(n):
    repere(-1.1,1.1,0.1,-1.1,1.1,0.1)
    #shape("turtle")
    width(2)
    up()
    goto(0,0)
    down()
    coef = sqrt(2)/2
    hypotenuse = 1
    angle = 0
    for k in range(n):
        setheading(angle)
        triangle(hypotenuse, k)
        hypotenuse *= coef
        angle += 45
    sauve("spirale")
    exitonclick()
```

```python
spirale(10)
```

## Tapis de Sierpinski

```python
#imports des modules pygame
import pygame
from pygame.locals import *


def sierpinski(left, top, width, height):
    
    #avec memoization des motifs c'est beaucoup, beaucoup plus rapide
    motifs = {}
    
    
    
    def sierpinski2(left, top, width, height):
        if (width, height) in motifs:
            motif_surf = motifs[(width, height)]
        elif width==1 or height == 1:
            motif_surf = pygame.surface.Surface((width*echelle, height*echelle))
            motif_surf.fill((255,255,255))
            pygame.draw.rect(motif_surf, (255,255,255), motif_surf.get_rect(),0)
            motifs[(width, height)] = motif_surf
        else:
            motif_surf = pygame.surface.Surface((width*echelle, height*echelle))
            dw, dh = width//2, height//2
            coin_surf = pygame.surface.Surface((dw*echelle, dh*echelle))
            coin_surf.fill((200,200,200))
            motif_surf.blit(coin_surf,(0*echelle, dh*echelle))
            pygame.draw.rect(motif_surf, (255,255,255), coin_surf.get_rect().move((0*echelle, dh * echelle)),1)
            for k in range(2):
                for j in range(2):
                    if (k,j) != (0,1):
                        motif_surf.blit(sierpinski2(k*dw * echelle, j*dh*echelle, dw, dh), (k*dw*echelle, j*dh*echelle))
            motifs[(width, height)] = motif_surf
        ecran.blit(motif_surf, (left*echelle, top*echelle))
        pygame.display.flip()
        return motif_surf
                    
    
    sierpinski2(left, top, width, height)
    return motifs
    
    
    
    
#initialisation des modules
pygame.init()

echelle = 2 ** 12

for exp in range(1, 12):
    echelle = echelle // 2
    L = 2**exp

    chrono = pygame.time.Clock()

    debut = chrono.tick()

    ecran = pygame.display.set_mode((L *echelle, L * echelle))
    #ecran.fill((0, 255, 0))

    motifs = sierpinski(0,0,L,L)
    pygame.image.save(ecran, 'carpette-2019-%sx%s.png'%(L, L))

    fin = chrono.tick()

    print('Temps : ', (fin - debut)/1000)
    
pygame.quit()
```
