```python
import matplotlib.pyplot as plt
from math import sqrt



##Graphique

def graphique(n, suite):
  x = list(range(n))
  y = [suite(k) for k in x]
  plt.clf()
  plt.plot(x, y,'b.')
  plt.title(f'{suite.__name__}-{n}-points')
  plt.savefig(f'{suite.__name__}-{n}-points.png')
  
```

```python
## Exemple 1  


def suite1(n):
  return 4 - 0.5 ** n


graphique(20, suite1)
```

```python


## Exemple 2

def suite2(n):
  return 4 + 10 * (-0.5) ** n


graphique(20, suite2)
```

```python
## Exemple 3

def suite3(n):
  u = 10
  for k in range(n):
    u = -0.5 * u + 1
  return u

graphique(20, suite3)
```

```python
## Exemple 4

def suite4(n):
  u = 100000
  for k in range(n):
    u = -0.5 * u + 1
  return u

graphique(20, suite4)
```

```python


## Exemple 5

def suite5(n):
  u = 0.0001
  for k in range(n):
    u = 2 * u + 1
  return u

graphique(20, suite5)
```

```python

## Exemple 6

def suite6(n):
  u = 10
  for k in range(n):
    u = sqrt(2 * u)
  return u

graphique(20, suite6)
```

```python

## Exemple 7

def suite7(n):
  u = 0.8
  for k in range(n):
    u = 2 * u * (1 - u)
  return u

graphique(20, suite7)

```
